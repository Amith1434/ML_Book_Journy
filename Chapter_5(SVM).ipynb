Support Vector Machines

Linear SVM Classification

```python
import numpy as np 
from sklearn import datasets
from sklearn.pipeline import Pipeline 
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:,(2,3)]    #importing only two columns
y = (iris["target"] == 2).astype(np.float64) #virginica

svm_clf = Pipeline([
    ("scaler",StandardScaler()),
    ("Linear_svc", LinearSVC(C=1,loss="hinge"))
])

svm_clf.fit(X,y)
```

```python
svm_clf.predict([[5.5,1.7]])
```

NonLinear SVM classification

```python
from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

X,y = make_moons(n_samples = 100,noise =.15, random_state =42) 
polynomial_svm_clf = Pipeline([
    ("poly_features",PolynomialFeatures(degree=3)),
    ("scaler",StandardScaler()),
    ("svm_clf",LinearSVC(loss="hinge"))
])

polynomial_svm_clf.fit(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
import numpy as np

x0_min, x0_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x1_min, x1_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(              ###Makes 1D array to 2D grid of all possible combinations
    np.linspace(x0_min, x0_max, 100),
    np.linspace(x1_min, x1_max, 100)
)
X_new = np.c_[xx.ravel(), yy.ravel()] ##Ravel is to flatten or reduce dimensions
y_pred = polynomial_svm_clf.predict(X_new)
y_pred = y_pred.reshape(xx.shape)

plt.contourf(xx, yy,y_pred,alpha=0.5,cmap="coolwarm")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Decision Boundary (SVM with Polynomial Features)")
plt.show()
```

Polynomial Kernel 

```python
from sklearn.svm import SVC
poly_kernel_svm_clf = Pipeline([
    ("scaler",StandardScaler()),
    ("svm_clf",SVC(kernel="poly",degree=3, coef0=1, C=1))
])

poly_kernel_svm_clf.fit(X,y)
```

```python
x0_min, x0_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x1_min, x1_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(              ###Makes 1D array to 2D grid of all possible combinations
    np.linspace(x0_min, x0_max, 100),
    np.linspace(x1_min, x1_max, 100)
)
X_new = np.c_[xx.ravel(), yy.ravel()] ##Ravel is to flatten or reduce dimensions
y_pred = poly_kernel_svm_clf.predict(X_new)
y_pred = y_pred.reshape(xx.shape)

plt.contourf(xx, yy,y_pred,alpha=0.5,cmap="coolwarm")
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="coolwarm", edgecolors="k")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Decision Boundary (SVM with Polynomial Features)")
plt.show()
```
